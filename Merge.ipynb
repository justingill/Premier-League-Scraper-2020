```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import warnings
```

```python
cwd = os.getcwd()
```

```python
standard = pd.read_csv(cwd + '\\processed_data\Standard_Stats_Table.csv')
goal_shot = pd.read_csv(cwd + '\\processed_data\Goal_Shot_Creation_Table.csv')
misc = pd.read_csv(cwd + '\\processed_data\Misc_Table.csv')
defensive_action = pd.read_csv(cwd + '\\processed_data\Defensive_Action_Table.csv')
adv_goalkeeping = pd.read_csv(cwd + '\\processed_data\Advanced_Goalkeeping_Table.csv')
goalkeeping = pd.read_csv(cwd + '\\processed_data\Goalkeeping_Table.csv')
pass_type = pd.read_csv(cwd + '\\processed_data\Pass_Types_Table.csv')
passing = pd.read_csv(cwd + '\\processed_data\Passing_Table.csv')
play_time = pd.read_csv(cwd + '\\processed_data\Play_Time_Table.csv')
possess = pd.read_csv(cwd + '\\processed_data\Possession_Table.csv')
```

```python
list_vars = [standard,goal_shot,misc,defensive_action,adv_goalkeeping,goalkeeping,pass_type,passing,play_time,possess]

merge = pd.DataFrame()

for i,var in enumerate(list_vars):
    if i == 0:
        merge = standard
    else:
        merge = merge.merge(var,how='left',on=['player','team','nationality'])
```

```python
merge.info()
```

```python
dropped = [each for each in merge.columns if each.endswith('_x')]
```

```python
dropped
```

```python
merge.drop(dropped,axis=1,inplace=True)
```

```python
merge.info()
```

```python
name_change = dict(zip([each for each in merge.columns if each.endswith('_y')],[each[:-2] for each in merge.columns if each.endswith('_y')]))
```

```python
merge.rename(name_change,axis=1,inplace=True)
```

```python
merge = merge.loc[:,~merge.columns.duplicated()]
```

```python
merge.info(verbose=True)
```

```python
merge.columns= merge.columns.str.lower()
merge = merge[sorted(merge.columns)]
```

```python
merge.info(verbose=True)
```

```python
merge.to_csv(cwd + '\\processed_data\Merged_Table.csv',index=False)
```
